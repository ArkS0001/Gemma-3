````python
from transformers import pipeline
import torch

pipe = pipeline(
    "image-text-to-text",
    model="google/gemma-3-4b-it",
    device="cuda",
    torch_dtype=torch.bfloat16
)
````

````python
messages = [
    {
        "role": "system",
        "content": [{"type": "text", "text": "You are Advanced Technical assistant."}]
    },
    {
        "role": "user",
        "content": [
            {"type": "image", "url": "/content/worked5.png"},
            {"type": "text", "text": "give complete detail of the flowchart"}
        ]
    }
]

output = pipe(text=messages, max_new_tokens=2048)
generated_text = output[0]['generated_text']

# Prepare markdown content
markdown_text = f"## Generated Response\n\n```\n{generated_text}\n```"

# Save the markdown content to a file named "output.md"
with open("output.md", "w") as f:
    f.write(markdown_text)

print("Markdown file saved as output.md")

````

````python
import ipywidgets as widgets
from IPython.display import display, HTML
import json

# Define input fields
text_input = widgets.Textarea(
    placeholder="Enter your query...",
    description="Text:",
    layout=widgets.Layout(width="100%", height="80px")
)

image_input = widgets.Text(
    placeholder="Enter image URL (optional)...",
    description="Image URL:",
    layout=widgets.Layout(width="100%")
)

output_area = widgets.Output()

# Function to process input and call `pipe()`
def generate_response(_):
    # Create messages format
    messages = [
        {
            "role": "system",
            "content": [{"type": "text", "text": "You are an Advanced Technical assistant."}]
        },
        {
            "role": "user",
            "content": []
        }
    ]

    # Add text input
    if text_input.value.strip():
        messages[1]["content"].append({"type": "text", "text": text_input.value.strip()})

    # Add image input if provided
    if image_input.value.strip():
        messages[1]["content"].append({"type": "image", "url": image_input.value.strip()})

    # Call the pipeline (Replace this with actual `pipe` model call)
    try:
        output = pipe(text=messages, max_new_tokens=1024)  # Call your model
        generated_text = output[0]['generated_text']
    except Exception as e:
        generated_text = f"Error: {str(e)}"

    # Display response
    with output_area:
        output_area.clear_output()
        display(HTML(f"<h4>Generated Response:</h4><pre>{generated_text}</pre>"))

# Button to trigger response generation
generate_button = widgets.Button(description="Generate")
generate_button.on_click(generate_response)

# Display UI elements
display(text_input, image_input, generate_button, output_area)
````

````python
from transformers import Qwen2_5_VLForConditionalGeneration, AutoTokenizer, AutoProcessor
from qwen_vl_utils import process_vision_info
import torch
# default: Load the model on the available device(s)
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2.5-VL-3B-Instruct", torch_dtype="auto", device_map="auto"
)

# # We recommend enabling flash_attention_2 for better acceleration and memory saving, especially in multi-image and video scenarios.
# model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
#     "Qwen/Qwen2.5-VL-3B-Instruct",
#     torch_dtype=torch.bfloat16,
#     attn_implementation="flash_attention_2",
#     device_map="cuda",
# )

# default processer
processor = AutoProcessor.from_pretrained("Qwen/Qwen2.5-VL-3B-Instruct")

# The default range for the number of visual tokens per image in the model is 4-16384.
# You can set min_pixels and max_pixels according to your needs, such as a token range of 256-1280, to balance performance and cost.
# min_pixels = 256*28*28
# max_pixels = 1280*28*28
# processor = AutoProcessor.from_pretrained("Qwen/Qwen2.5-VL-3B-Instruct", min_pixels=min_pixels, max_pixels=max_pixels)

messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "image": "/content/gemmatest3.png",
            },
            {"type": "text", "text": "Describe this image."},
        ],
    }
]

# Preparation for inference
text = processor.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)
image_inputs, video_inputs = process_vision_info(messages)
inputs = processor(
    text=[text],
    images=image_inputs,
    videos=video_inputs,
    padding=True,
    return_tensors="pt",
)
inputs = inputs.to("cpu")

# Inference: Generation of the output
generated_ids = model.generate(**inputs, max_new_tokens=128)
generated_ids_trimmed = [
    out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
]
output_text = processor.batch_decode(
    generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
)
print(output_text)
````
